## Funciones basicas a implementar

Implemente las siguientes funciones.
- Funcion de activacion sigmoide

$$\sigma(x) = \frac{1}{1+e^{-x}}$$

- Prediccion 

$$\hat{y} = \sigma(w_1 x_1 + w_2 x_2 + b)$$

- Funcion error

$$Error(y, \hat{y}) = - y \log(\hat{y}) - (1-y) \log(1-\hat{y})$$

- Funcion para actualizar los pesos

$$ w_i \longrightarrow w_i + \alpha (y - \hat{y}) x_i$$

$$ b \longrightarrow b + \alpha (y - \hat{y})$$

```python
# Activation (sigmoid) function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

#  funcion Sigmoide

Esta función es la función de activación sigmoidal. Toma un número x como entrada y devuelve el resultado de la función sigmoidal en x, que está en el rango (0, 1). La función sigmoidal se utiliza para mapear cualquier valor de entrada a un valor entre 0 y 1 y se utiliza en la regresión logística para clasificar las instancias
 

```python
# Output (prediction) formula
def prediccion(features, weights, bias):
    return sigmoid(np.dot(features, weights) + bias)
```

#  funcion Prediccion

Esta función es la fórmula de salida de la regresión logística. Toma un conjunto de características features, un conjunto de pesos weights y un sesgo bias como entrada y devuelve la predicción de la regresión logística basada en estas entradas. La fórmula es simplemente la activación sigmoidal aplicada al producto punto de las características y pesos más el sesgo.

```python
# Error (log-loss) formula
def error_formula(y, output):
    return np.dot(-y, np.log(output)) - np.dot((1-y), np.log(1-output))
```

# Funcion Regresion logistica
Esta función es la fórmula de error utilizada para evaluar la precisión de las predicciones de la regresión logística. Toma las etiquetas verdaderas y y las predicciones output como entrada y devuelve el error de log-loss. La fórmula de log-loss se utiliza comúnmente en la clasificación binaria para calcular la diferencia entre las predicciones del modelo y las etiquetas verdaderas.

```python
# Gradient descent step
def update_weights(x, y, weights, bias, learnrate):
    output = prediccion(x, weights, bias)
    weights += np.dot(learnrate, np.dot((y - output),x))
    bias += learnrate*(y- output)
    return weights, bias
```

# Funcion Gradiente decendiente
Esta función realiza un paso de descenso de gradiente estocástico en la regresión logística. Toma un conjunto de características x, un conjunto de etiquetas verdaderas y, un conjunto de pesos weights, un sesgo bias y una tasa de aprendizaje learnrate como entrada y actualiza los pesos y el sesgo utilizando la regla de actualización de descenso de gradiente estocástico. La función prediccion() se utiliza para calcular las predicciones y luego los pesos y el sesgo se actualizan utilizando la fórmula de actualización de descenso de gradiente estocástico. La función devuelve los nuevos pesos y el sesgo actualizados.

## Funcion de entrenamiento
Esta funcion hace una interacion del algoritmo del gradiente descendente en todos los datos, para un numero determinado de epocas y algunas lineas limite a medida que avanza el algoritmo.

```python
np.random.seed(44)

epochs = 100
learnrate = 0.01

def train(features, targets, epochs, learnrate, graph_lines=False):
    
    errors = []
    n_records, n_features = features.shape
    last_loss = None
    weights = np.random.normal(scale=1 / n_features**.5, size=n_features)
    bias = 0
    for e in range(epochs):
        del_w = np.zeros(weights.shape)
        for x, y in zip(features, targets):
            output = prediccion(x, weights, bias)
            error = error_formula(y, output)
            weights, bias = update_weights(x, y, weights, bias, learnrate)
        
        # Printing out the log-loss error on the training set
        out = prediccion(features, weights, bias)
        loss = np.mean(error_formula(targets, out))
        errors.append(loss)
        if e % (epochs / 10) == 0:
            print("\n========== Epoch", e,"==========")
            if last_loss and last_loss < loss:
                print("Train loss: ", loss, "  WARNING - Loss Increasing")
            else:
                print("Train loss: ", loss)
            last_loss = loss
            predictions = out > 0.5
            accuracy = np.mean(predictions == targets)
            print("Accuracy: ", accuracy)
        if graph_lines and e % (epochs / 100) == 0:
            display(-weights[0]/weights[1], -bias/weights[1])
            

    # Plotting the solution boundary
    plt.title("Solution boundary")
    display(-weights[0]/weights[1], -bias/weights[1], 'black')

    # Plotting the data
    plot_points(features, targets)
    plt.show()

    # Plotting the error
    plt.title("Error Plot")
    plt.xlabel('Number of epochs')
    plt.ylabel('Error')
    plt.plot(errors)
    plt.show()
```

- Un plot de la funcion error. Note que a medida que avanza las epocas, esta disminuye.

```python
train(X, y, epochs, learnrate, True)
```
